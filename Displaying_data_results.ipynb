```python
import pandas as pd
import plot_likert

file_path = "./data/wellbeing_survey.csv"

try:
    # Read the CSV file
    df = pd.read_csv(file_path)

    # Display the first 10 rows of the data to check format, DEBUG
    #print("Survey Results (first 10 rows):")
    #print(df.head(10))

    # Check the unique values to identify responses
    questions = [
        "I felt happy.",
        "I felt engaged.",
        "I felt comfortable.",
        "I felt safe and secure.",
        "I enjoyed the company of other people.",
        "I talked to other people.",
    ]

    # Get relevant columns for Likert scale analysis
    likert_data = df[questions]
    unique_responses = pd.unique(likert_data.values.ravel('K'))
    print("Unique responses found in the data:", unique_responses) #Debug

except FileNotFoundError:
    print(f"Error: The file at {file_path} was not found.") 
except pd.errors.EmptyDataError:
    print("Error: The CSV file is empty.")
except Exception as e:
    print(f"An error occurred: {e}")
     
```

```python
# use agree5 to plot the data
display(likert_data)
myscale = ("Strongly agree", "Agree", "Neither agree nor disagree", "Disagree", "Strongly disagree")
plot_likert.plot_likert(likert_data, myscale, plot_percentage=True, figsize=(11,2));
```

```python
# with percentages for each level
plot_likert.plot_likert(likert_data, myscale, plot_percentage=True,
                        bar_labels=True, bar_labels_color="snow", colors=plot_likert.colors.default_with_darker_neutral, figsize=(11,2));
```
